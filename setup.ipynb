# setup.ipynb

This script creates a venv and sets up required packages. Nothing fancy.

macOS/Linux only for now.

```python
import os
import platform
import subprocess
import sys

def setup_uv_environment():
    system = platform.system()
    print(f"Detected OS: {system}")

    # 1. Install uv based on OS
    try:
        if system == "Windows":
            print("Installing uv via PowerShell...")
            cmd = 'powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"'
            subprocess.run(cmd, shell=True, check=True)
            # Add to path for current session
            uv_path = os.path.expanduser("~\\.cargo\\bin")
        else:
            print("Installing uv via cURL...")
            cmd = "curl -LsSf https://astral.sh/uv/install.sh | sh"
            subprocess.run(cmd, shell=True, check=True)
            # Add to path for current session
            uv_path = os.path.expanduser("~/.local/bin")

        # Update the current process PATH
        if uv_path not in os.environ["PATH"]:
            os.environ["PATH"] = f"{uv_path}{os.pathsep}{os.environ['PATH']}"
            
        print("uv installed successfully.")

        # 2. Install the requested packages
        packages = ["numpy", "matplotlib", "torch", "torchsde", "scipy", "pandas", "openap"]
        print(f"Installing packages: {', '.join(packages)}...")
        
        # Use 'uv pip install' for immediate notebook use
        subprocess.run(["uv", "pip", "install", "--system"] + packages, check=True)
        
        print("\n" + "="*30)
        print("SETUP COMPLETE")
        print("="*30)

    except subprocess.CalledProcessError as e:
        print(f"Error during installation: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    setup_uv_environment()
```
